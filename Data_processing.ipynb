```python
import re

genres_clean = []

for i in range(len(df_movies_metadata)):
    genre = df_movies_metadata.iloc[i,3]
    try:
        genres_clean.append(', '.join(pd.DataFrame.from_dict(eval(genre[1:-1]))['name'].sort_values(ascending=True)))
    except:
        try:
            if len(genre) >= 3:
                genres_clean.append(re.findall('\'name\': \'([\w\s]*)\'',genre)[0])
            else:
                genres_clean.append('')
        except:
            print(f'Error {i}: {genre}')
    
len(genres_clean)
```
